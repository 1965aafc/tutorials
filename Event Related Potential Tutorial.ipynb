```python
#loading Modules
%config InlineBackend.figure_format = 'retina'
%pylab inline
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
import scipy.io
import scipy.signal
```

```python
#loading EEG data
filename = 'emodat.mat'
filename = os.path.join('/Users/Torben/Desktop/Voytek/', filename)
data = sp.io.loadmat(filename)
srate = data['srate'] # sampling rate
data = data['data'] # time series
data = data[0, :]
print srate
```

With an event related potential, we're looking at what happens to the field potential at certain time indices. 
Often we're interested in what participants' brains do around the times when we present stimuli.
i.e. We might want to see what participants brains do when we show them one type of stimulus versus another.

Here's how we can calculate it:

```python
#A super basic way of doing an ERP is looking at the window surrounding a single event like this:
event_inds = [1034, 2133, 4681, 9960, 12574]
this_event = event_inds[2]

my_ERP = [data[this_event-100:this_event + 100]]

plot(squeeze(my_ERP))
```

This tells us a little bit about about what someone's brain does in a certain time period 
but this is only one event in one channel. We can do better by getting multiple events 
(and eventually multiple channels and participants) involved in our analysis.

Now let's do a bigger and better method of computing ERP:

```python
#First we need to define a time window around the event that we're examining (In milliseconds).

before = -100 #100 ms before event
after = 500 #500 ms after

#account for sampling rate
before_ind = int(round((float(before) / 1000) * srate));
after_ind = int(round((float(after) / 1000) * srate));

#Here we define the indices (in ms) of the events that we are interested in -> when stimuli are presented


#We also need some 'standard' voltage data to compare to the responses to the targets. We do this by defining 
#a time window (in ms) before the event that we will get some non-target voltage data. I'll explain this more later.
baselineWindow = [-100,0] 

print before_ind
print after_ind
```

```python
#initializing ERP variables. 
ERP_mean = [0] *int(floor(abs(before_ind) + abs(after_ind) + 1))
size_of_mean = size(ERP_mean)
print size_of_mean

size_of_event_inds = size(event_inds)

#initializing baseline time window and matrix
baselineVals =  [0]*abs(baselineWindow[1]-baselineWindow[0]) 
base_mat = [[0 for k in baselineVals] for j in xrange(size_of_event_inds)]
print size(base_mat)

#intitializing ERP matrix
ERP_mat = [[0 for k in xrange(size_of_mean)] for j in xrange(size_of_event_inds)]
print size(ERP_mat)

#initializing ERP percent change matrix
ERP_prcMat = [[0 for k in xrange(size_of_mean)] for j in xrange(size_of_event_inds)]
```

Here we're getting all of the time indices for each window around our events using the ranges we created above.
For each time index we get the correspoinding data point.
This produces a matrix with each row representing an event and each column representing a time in the window 
surrounding each event

```python
for x in range(size(event_inds)):
    #get time points of data we want
    ERP_mat[x]=[range(event_inds[x]- abs(before_ind), event_inds[x] +abs(after_ind)+1)]
    #Fill in data using time points
    temp = []
    for y in ERP_mat[x]:
        temp.append(data[y])
        ERP_mat[x] = temp

#Now we have a matrix that looks like this:    
#event 1 | voltage at timepoint 1a | voltage at timepoint 1b | voltage at timepoint 1n | voltage at timepoint 1 when target is displayed | voltage at timepoint 1x | voltage at timepoint 1y |...
#event 2 | voltage at timepoint 2a | voltage at timepoint 2b | voltage at timepoint 2n | voltage at timepoint 2 when target is displayed | voltage at timepoint 2x | voltage at timepoint 2y |...
#event n | voltage at timepoint 3a | voltage at timepoint 3b | voltage at timepoint 3n | voltage at timepoint 3 when target is displayed | voltage at timepoint 3x | voltage at timepoint 3y |...
            
      
#fill data in baseline Window before each event
for x in range(size(event_inds)):
    base_mat[x]=[range(event_inds[x]- abs(baselineWindow[0]), event_inds[x] +abs(baselineWindow[1])+1)]
    temp = []
    for y in base_mat[x]:
        temp.append(data[y])
        base_mat[x] = temp
    
```

Getting the mean ERP from ERP matrix
The mean ERP is just the average voltage of each time point across trials. 
i.e. We take the average of each column in the ERP matrix
Now we know what what the brain tends to do before and after events are displayed.

```python
ERP_mean = np.squeeze(np.mean(ERP_mat,0))
plot(ERP_mean)
#when is the target presented on this graph? If you type print ERP_mean[target_time] it should return -3.77980004046
```

Getting the mean baseline from baseline matrix
This is just like the ERP matrix 

```python
base_means = np.squeeze(np.mean(base_mat,2))
print base_means
```

We can assume that these values represent the inherent noise in the system around each event presentation. 
As you can see there is a different ammount of noise around each event so we should maybe take note of this
(but when we subtract out all of the noise for an average time window, these values will average out)
Because there is inhearent noise, the above graph doesn't accurately represent the actual voltage changes 
in our time windows.

```python
#Subtracting noise from the ERP matrix
temp_mat = [[0 for k in xrange(size_of_mean)] for j in xrange(size_of_event_inds)]
for x in range(size(event_inds)):
    temp_mat[x]=[y - base_means[x] for y in ERP_mat[x]]
```

Plotting ERP accounting for noise vs. no noise.

```python
adjusted_ERP_mean = np.squeeze(np.mean(temp_mat,0))
plot(ERP_mean,color ='b')
plot(adjusted_ERP_mean, color ='r')
#Cool. We subtracted out the noise and everything shifted up.
```

Now let's look at how the voltage changes as a percent increase from baseline

```python
for x in range(size(event_inds)):
    ERP_prcMat[x]=[y - base_means[x] for y in ERP_mat[x]]
print size(ERP_prcMat[1])    
ERP_prc = np.squeeze(np.mean(ERP_prcMat,0))

plot(ERP_prc)
```

That's it! In Summary here's what we learned:
1.  You can look at the time window surrounding a single event or multiple events to see what is going on
    An event in an ERP doesn't have to be a stimulus presentation. An event can be whatever you want
    such as peaks of theta waves or when the participant blinks
2. If you want an accurate ERP, you have to subtract out the baseline noise
3. You can also look at ERP's as a percent increase from baseline if you want.
