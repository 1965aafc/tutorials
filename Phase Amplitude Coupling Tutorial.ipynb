When we look at the phase amplitude coupling in some EEG data, we're investigating how much 
the phase of a low frequency wave (Theta  - 3-8 Hz) influences the amplitude of a high 
frequency wave (Gamma - 80-150 Hz). If that definition isn't intuitive now, it will make
more sense later on

First, we're going to need some functions that isolate the frequencies that we're interested in.
This function gives us the parameters that we'll need to apply the filter. The next uses those
parameters to actually filter the data.

```python
def butter_bandpass(lowcut, highcut, fs, order=4):
    #lowcut is the lower bound of the frequency that we want to isolate
    #hicut is the upper bound of the frequency that we want to isolate
    #fs is the sampling rate of our data
    nyq = 0.5 * fs #nyquist frequency - see http://www.dspguide.com/ if you want more info
    low = float(lowcut) / nyq
    high = float(highcut) / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

#This function applies Butterworth filter to data
def butter_bandpass_filter(mydata, lowcut, highcut, fs, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, mydata)
    return y
```

This function is a quick way for us to calculate ERP. Read the ERP walkthrough if you're interested

```python
def ERP(data, srate, event_inds, before, after, baselineWindow = 0):
    #define time window
    before_ind = int(round((float(before) / 1000) * srate))
    after_ind = int(round((float(after) / 1000) * srate))
    #initialize ERP
    ERP_mean = [0] *int(floor(abs(before_ind) + abs(after_ind) + 1))
    size_of_mean = size(ERP_mean)
    size_of_event_inds = size(event_inds)
    
    #intitialize ERP Matrix
    ERP_mat = [[0 for k in xrange(size_of_mean)] for j in xrange(size_of_event_inds)]
    #fill ERP_matrix with equal time epochs
    for x in range(size(event_inds)):
        ERP_mat[x]=[range(event_inds[x]- abs(before_ind), event_inds[x] +abs(after_ind)+1)]
        #Fill data into time epochs
        temp = []
        for y in ERP_mat[x]:
            temp.append(data[y])
            ERP_mat[x] = temp
    #fill data in baseline Window before each event
    if baselineWindow is not 0:
        #initialize baseline time window
        baselineVals =  [0]*abs(baselineWindow[1]-baselineWindow[0]) 
        base_mat = [[0 for k in baselineVals] for j in xrange(size_of_event_inds)]
        for x in range(size(event_inds)):
            base_mat[x]=[range(event_inds[x]- abs(baselineWindow[0]), event_inds[x] +abs(baselineWindow[1])+1)]
        temp = []
        for y in base_mat[x]:
            temp.append(data[y])
            base_mat[x] = temp
        base_means = np.squeeze(np.mean(base_mat,2))
    
        #subtract baseline from ERP_mat
        for x in range(size(event_inds)):
            ERP_mat[x]=[y - base_means[x] for y in ERP_mat[x]]
    #subtract baseline from mean 
    ERP_mean = np.squeeze(np.mean(ERP_mat,0))
    ERP_mat = np.squeeze(ERP_mat)
    #everything = [adjusted_ERP_mean, adjusted_ERP_mat, ERP_prc]
    return ERP_mean, ERP_mat
```

In the graph above, we have two functions, the phase of theta (blue )and the amplitude of gamma (green). 
The phase line shows what phase the theta band is (in radians) in our time domain. 
It oscillates between pi and negative pi.
We also have the envelope surrounding the amplitude of our gamma wave. It seems like there might be some relationship
between the two. We use Phase Amplitude Cupling (PAC) to measure that relationship.

We can't do a simple correlation to see how the two functions relate
so we're going to make a function that correlates periodic data with linear data. 

```python
def circCorr(ang,line):
    n = len(ang)
    rxs = sp.stats.pearsonr(line,sin(ang))
    rxs = rxs[0]
    rxc = sp.stats.pearsonr(line,cos(ang))
    rxc = rxc[0]
    rcs = sp.stats.pearsonr(sin(ang),cos(ang))
    rcs = rcs[0]
    rho = sqrt((rxc**2 + rxs**2 - 2*rxc*rxs*rcs)/(1-rcs**2)) #r
    r_2 = rho**2 #r squared
    pval = 1- sp.stats.chi2.cdf(n*(rho**2),1)
    standard_error = sqrt((1-r_2)/(n-2))

    return rho, pval, r_2,standard_error

plot(phase_data[1:1000])
plot(amp_data[1:1000])

print circCorr(phase_data[1:1000], amp_data[1:1000])

#The below values represent the r, p, r^2 and standard of error respectively
#The values that we got show that there seems to be a small but significant correlation. Cool.
```

In this module, we learned that we can examine the relationship between the phase of a slow wave 
and the amplitude of a faster wave.
To isolate the frequency ranges that we want, we use filters then do some math to get the phase
and amplitude of the waves that we're interested in.
Then we can use statistics to quantify the relationship between the two.
In this example we have shown that there is a small but significant correlation.

PAC is useful because we can compare the PAC at different time windows in a behavioral task or
compare PAC between individuals. This is a relatively new technique and there are probably several
more ways to draw conclusions about neural activity using PAC.
