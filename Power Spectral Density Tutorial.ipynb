In this tutorial we're going to look at PSD. The PSD of a signal tells us which **frequencies** are most present within a **time** window.

```python
from __future__ import division

#loading modules
%config InlineBackend.figure_format = 'retina'
%pylab inline
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
import scipy.io
import scipy.signal
```

```python
#loading data
filename = 'emodat.mat'
filename = os.path.join('/Users/Torben/Desktop/Voytek/', filename)

data = sp.io.loadmat(filename)
srate = data['srate']
data = data['data'] #ECoG data
data =  squeeze(data); #should be 1459920
print srate #should be 1017.25333333
print len(data) #should be 1476035 
```

The sampling rate is important for how we analyze the data. When you digitally record a signal, you measure the output at a **certain rate**. In this case we gathered 1017.25333333 samples of data each second. If we want, we can **resample** the data using a different sampling rate. 

**Resampling, PSD**, and many other things we do in this lab fall under the title of Digital Signal Processing. Read about it for free here:
http://dspguide.com/

In the next cell we're going to resample our data to a different sampling rate. We're doing this beacuse it will help us to have an even number of samples in each second later on.


```python
newsrate = 1024 # new sampling rate

if srate!=newsrate: # resample
    data_resampled = sp.signal.resample(data, int(np.floor(len(data)*(newsrate/srate)))) 

#How many seconds of data do we have?
print len(data_resampled)/1024 #Total samples of data sampled at 1024 / sampling rate = total seconds sampled

#so we actually have 1451 seconds of data. Cool.
```

So there's this complicated math function called the **Fourier transform.**
This function measures the **power** of each **frequency** in a signal. The math that it takes to do this is very challenging.

Imagine you play a chord on a piano. If you ran a Fourier transform on the sound signal, you could see how hard you pressed each of the keys.

In the brain, there are a lot of frequencies present in the same signal and they aren't as simple as a chord with 3 discrete notes.

Let's look at a complex signal



```python
f = 1024 #sampling frequency
dur = 10 #10 seconds of signal
freq = 10 #10 Hz signal
freq2 = 130 #5- Hz signal
t = arange(0, dur, 1/f) #times for d
sig1 = sin(2 * pi * freq * t) #10 Hz wavelength
sig1 = 1.5*sig1; #increase the power of signal 1
sig2 = sin(2 *pi * freq2 * t) #130 Hz wavelength
#plt.plot(t, sig1, label = 'Signal 1')
#plt.plot(t, sig2, label = 'Signal 2')
plt.plot(t[0:512],sig1[0:512]+sig2[0:512], label = 'complex signal') #plot 0.5 seconds of data

legend()
```

So in the graph above we can see that there are 2 wavelengths that seem to be important to this signal. There is a slower wave and a fast wave. We can use the fourier transform to pick out those frequencies.

We'll calculate the Fourier transform of the above signal in the cell below using the fast fourier transform (**fft**) method.

```python
fourier = np.fft.fft(sig1+sig2,1024)
plot(abs(fourier[0:150]))
plt.ylabel('Power')
plt.xlabel('Frequency (Hz)')
plt.title('FFT of a complex signal')
```

So in this plot we're looking at the power of all of the freqencies present in our signal. 

We can clearly see that there is a 10 Hz wavelength and a 130 Hz wavelength.

We can also see that the 10 Hz wavelength has a **greater amplitude** than the 130 Hz

Notice that when we do this, we lose **all** information about the **timing** in our signal. Now we know nothing about the **offset of the phases** of these signals.

Also, you'll see that I took the absolute value of only the first 160 time points of the fft. If you want to see why I did that, check www.dspguide.com


#####Now what does the fft of our complicated ECoG signal look like?


```python
fourier = np.fft.fft(data,1024)
plot(abs(fourier[0:150]))

plt.ylabel('Power')
plt.xlabel('Frequency (Hz)')
plt.title('FFT of ECoG signal')
```

Woah. So there's a lot going on. Let's break this down.

* This signal is **very** noisy. There appear to be **many** wavelengths that contribute to this signal.
* Longer wavelengths seem to have a **larger power** than shorter wavelengths. 

 

*Side Note*

Actually, the equation for this line (ignoring the noise) is P = 1/f. This is a very special mathmatical ratio called a **power law distribution** or **pink noise**. 1/f relationships show up **everywhere** in nature in spooky ways. Sizes of earthquakes, sizes of craters on the moon, and even the power spectrum of popular music **all have power law distributions**. Really interesting and spooky stuff.

####Back to neuroscience

There's a way to remove the noise from our fft without losing information about the wavelengths that are present. We do this using Welch's method.

Basically, we calculate the fft of a signal using a sliding window. Then we calculate the mean PSD from all of of the sliding windows.

####See below

```python
f,pspec = sp.signal.welch(data, fs=newsrate, window='hanning', nperseg=2*newsrate, noverlap=newsrate/2, nfft=None, detrend='linear', return_onesided=True, scaling='density')
#Try to figure out what the paramaters above are doing. What happens if you change them?

plot(f[0:150],log10(pspec[0:150]))
#Any frequencies with >150 Hz are going to be noise in ECoG data.
plt.ylabel('Power')
plt.xlabel('Frequency (Hz)')
plt.title("Welch's PSD of ECoG signal")
```

Look at that! 

Now the signal is much cleaner and there seem to be some interesting peaks around 8, 14, and 25 Hz. These are known as the theta, alpha, and beta bands.

These wavelengths have been studied since the 1950's but we really don't know much about them.
