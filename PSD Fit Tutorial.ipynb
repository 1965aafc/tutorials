The sampling rate is important for how we analyze the data. When we record EEG data we don't get every single instance. In this case we gathered 1017.25333333 samples of data per second. If we want, we can resample the data using a different sampling rate. 

This (and many other things we do in this lab) fall under the title of Digital Signal Processing. Read about it for free here:
http://dspguide.com/

In the next cell we're going to resample our data to a different sampling rate


```python
#newsrate is the new sampling rate that we want.
newsrate = 1024 # new sampling rate

if srate!=newsrate: # resample
    data = sp.signal.resample(data, int(np.floor(len(data)*(newsrate/srate)))) 
#floor is just a round-down-to-the-nearest-integer function

num_of_samps = 1476035/1024

print(num_of_samps)
#so we actually have 1441 seconds of data. Cool.
```

Here's the good stuff. We're going to actually calculate the power spectral density (PSD)of the data.
just kidding. We're going to use a function that scipy has already built in.

PSD shows how the power of the signal is distributed over a range of frequencies
We don't want to do all of the math so we're just goint to use Welch's method of calculating power spectral density.
look up how all of this works here if you're interested: http://docs.scipy.org/doc/scipy-dev/reference/generated/scipy.signal.welch.html


```python
f,pspec = sp.signal.welch(data, fs=newsrate, window='hanning', nperseg=2*newsrate, noverlap=newsrate/2, nfft=None, detrend='linear', return_onesided=True, scaling='density')

fwin = [0, 150] 
# this is the  range of frequencies that we're interested in.
#any frequencies with >150 Hz are going to be noise.

fwin = [np.argmin(abs(f-fwin[0])), np.argmin(abs(f-fwin[1]))] # pspec over freq range of interest
f = f[fwin[0]:fwin[1]]
#f is just a variable for the frequency range of interest in 0.5 Hz increments 

#plotting the PSD
pspec = pspec[fwin[0]:fwin[1]]
pspec = np.log10(pspec)
plt.plot(f,pspec)
```

Let's get some information about the spectrum we just plotted
specifically, we're interested in the function that describes this shape and the slope

```python
slope_params = np.polyfit(f, pspec, 2) # fit the curve of the PSD to isolate the oscillatory components
#np.polyfit gets vector components that minimize r^2 (polynomial function closest to all points)
slopefit = np.polyval(slope_params, f)
#fitting a 2nd degree polynomial to a graph of pspec using the paramaters defined by polyfit
# y = 5.14347266e-06 * x^2 -1.21294158e-02 * x + 6.21383500e-01

flatspec = pspec-slopefit # subtract out the slope fit from the spectrum to flatten it out
#pretty neat

plt.plot(f, pspec)
plt.plot(f, slopefit)
plt.plot(f, flatspec)
plt.legend(['original psd', 'PSD slope fit', 'flatspec (psd - 1/f)'], loc=1)
```

```python
dy = np.diff(slopefit) / np.diff(np.log10(f)) # derivative of slopefit over derivative of x axis scale
#finding the derivative to get the knee of the graph
#help(np.diff)

idx = np.argmin(abs(dy[0:(len(dy)/2)]-(-1))) #  trig to find the knee
knee = f[idx] # where's the knee?

plt.semilogx(f, slopefit)
plt.semilogx(f[idx], slopefit[idx], 'or')
plt.plot(dy)
plt.legend(['PSD slope fit', 'knee (tan == -1)','dy'], loc=3)
```

To describe the shape of the PSD better than a best fit line we want to find the peaks of the low frequency wave bands.


```python
peakind = sp.signal.find_peaks_cwt(flatspec, np.arange(0.1,10), wavelet=None)
#cool scipy function to find the peaks

peakind = np.array(peakind)
peakind = peakind[f[peakind]<=40]
#we only want the x values of peaks in the range we expect to see them i.e. the first 3 big ones
peakvals = flatspec[peakind]
#gets y vals of the peaks we found

plt.plot(f, flatspec)
plt.plot(f[peakind], flatspec[peakind], 'or')
plt.legend(['flatspec (psd - 1/f)', 'putative peaks'], loc=1)
```

Now we know where the peaks are and we have a pretty good estimate for the slope but now we have to fill in the peaks. We can do this by assuming that the peaks are gausian in nature and draw a line over each distribution.

```python

def norm(x, *args):
    number_of_gaussians = int(args[-1]) # number of gaussians from peak detection
    fitdat = np.array(0)
    for peaki in range(number_of_gaussians):
        fitdat = fitdat + args[peaki+(2*number_of_gaussians)]*sp.stats.norm.pdf(x, loc=args[peaki], scale=args[peaki+number_of_gaussians]) # add a bunch of gaussians together
    return fitdat
```

```python
#fit combination of gausians (norm function) to data
params = list(f[peakind]) + [2, 2, 4] + list(peakvals) + [len(peakind)] # initial guess for parameters based on peak detection
oscillatory_fit_params,_ = sp.optimize.curve_fit(norm, f, flatspec, p0=params) # fitting parameters
oscillatory_fit = norm(f, *oscillatory_fit_params) # getting gausian values that represent the peaks


plt.plot(f, flatspec)
#Flat power spectrum (because we subtracted the slopefit from each point)
plt.plot(f, oscillatory_fit)
#lost
plt.plot(f, pspec)
#vanilla power spectral density
plt.plot(f, pspec-oscillatory_fit)
#psd minus first 3 big oscillation peaks
plt.legend(['flatspec (psd - 1/f)','oscillatory fit', 'original psd', 'no oscillations'], loc=1)
```

```python
plt.plot(f, pspec)
plt.plot(f, (slopefit + oscillatory_fit))
#Cool now we have a fit to the curve that accounts for 3 oscillation peaks
plt.legend(['original spectrum', 'fitted spectrum'])
```
